```python
# Script that prints final results for paper


#Load libraries and dataset
import numpy as np
import pandas as pd
import math


import os
import time
#os.chdir('/Users/mr54725/Dropbox/Research/BellLabs_pathloss/terrain/')
from models import *
import pickle

from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt

# Importing gc module
import gc

#Load dataset
df = np.load('data/all_dataset2.npy')
columns = ['ht', 'loc_x', 'loc_y', 'measPathGain', 'profile', 'ranges',
       'terrainHtFile', 'curv', 'terrainDev', 'predPathGain', 'dist', 'angles',
       'cv_group', 'profile_sc']
df = pd.DataFrame(df, columns=columns)
numericas = ['ht', 'loc_x','loc_y', 'measPathGain','terrainHtFile','curv', 'terrainDev', 'predPathGain', 'dist', 'angles']
df = df.astype({col: np.float32 for col in numericas})
```

```python
# Slope intercept predictions

tot_folds = 10
n_obs  = df.shape[0]
n_test = np.ceil(n_obs/tot_folds)

predictions = []
cvscores = []
trainScores = []
for fold in range(tot_folds):
    if fold<tot_folds-1:
        test = range(int(fold*n_test), int((fold+1)*n_test))
    else:
        test = range(int(fold*n_test), n_obs)
    result = np.polyfit( np.delete(df.dist.values, test,0), np.delete(df.measPathGain.values, test,0), deg = 1)
    pred=result[1] + result[0]*df.dist.iloc[test]
    predictions.append(pred)
    predTrain = result[1] + result[0]*np.delete(df.dist.values, test,0)
    trainScores.append(np.sqrt(np.mean(np.power(np.delete(df.measPathGain.values,test,0)-predTrain, 2))))
    cvscores.append(np.sqrt(np.mean(np.power(df.measPathGain.iloc[test]-pred, 2))))
print( 'Mean cross-validation RMSE:   '+ str(np.mean(cvscores)))

print('mean gap :  '+ str(np.mean(np.array(cvscores) - np.array(trainScores))) ) 
```
