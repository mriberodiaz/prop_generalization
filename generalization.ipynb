```python
# This script shows how some models where trained... 

import numpy as np
import pandas as pd
import math


import os
import time
from models import *
import pickle

from keras.preprocessing.sequence import pad_sequences
from keras.callbacks import ModelCheckpoint

# Importing gc module
import gc

```

```python
df = np.load('data/all_dataset2.npy')
columns = ['ht', 'loc_x', 'loc_y', 'measPathGain', 'profile', 'ranges',
       'terrainHtFile', 'curv', 'terrainDev', 'predPathGain', 'dist', 'angles',
       'cv_group', 'profile_sc']
df = pd.DataFrame(df, columns=columns)
numericas = ['ht', 'loc_x','loc_y', 'measPathGain','terrainHtFile','curv', 'terrainDev', 'predPathGain', 'dist', 'angles']
df = df.astype({col: np.float32 for col in numericas})
```

```python
##################################################################
##################################################################
#######     TRY Again old models without regularization     ######
##################################################################
##################################################################


descriptions = ['dense1_', 'dense2_','dense3_','dense4_','dense5_','dense6_','dense7_','dense8_','dense9_']
models  = [dense_1(), dense_2(),dense_3(),dense_4(),dense_5(),dense_6(),dense_7(),dense_8(),dense_9()]

X = pad_sequences(df.profile_sc)
X = np.append(X, np.expand_dims(df.dist.values, axis = 1), axis = 1)
y = np.array(df.measPathGain)

n_obs  = X.shape[0]
n_test = np.ceil(n_obs/5)



for i in range(9):
    predictions = []
    print(' \n Started training \n ')
    print(' \n Model:    ' + str(i+1))
    print('\n')
    t0 = time.time()
    cvscores = []
    for fold in range(5):
        
        model = models[i]
        desc = descriptions[i]+ str(fold) + '_' 

        filepath="models/" + desc + time.strftime('%l_%M%p_%b_%d_%Y')[1:] + "-{epoch:02d}-{val_loss:.2f}.hdf5"
        callback = ModelCheckpoint(filepath, monitor='loss', verbose=0, save_best_only=True)
        if fold<4:
            test = range(int(fold*n_test), int((fold+1)*n_test))
        else:
            test = range(int(fold*n_test),n_obs )
            
    
        tf1 = time.time()
        history  = model.fit(np.delete(X, test,0), np.delete(y, test,0), 
                            validation_data = (X[test], y[test]), 
                            epochs=300, batch_size=32, 
                            callbacks = [callback],
                            verbose=0)
        pickle.dump(history.history, open('resultsCV/history_'+desc+ '.p', 'wb'))
        pred = model.predict(X[test])
        predictions.append(pred)
        scores = model.evaluate(X[test], y[test], verbose=0)
        cvscores.append(scores)
        tf2 = time.time()
        print('finished fold:  ' + str(fold) + '  -  Time:  ' + str(tf2-tf1) + '  -  score:  ' + str(scores) )
        gc.collect()

    pickle.dump(predictions, open('resultsCV/predictions_'+desc+ '.p', 'wb'))
    tf = time.time()
    time_file = np.array([t0,tf,tf-t0])
    np.save('resultsCV/' +desc+'_cvresults.npy',cvscores)
    np.save('resultsCV/' +desc+'_time.npy',time_file)
    print(' \n CV mean rmse:  '+ str(np.mean(np.sqrt(cvscores)))+ '  +/-  ' +str(np.std(np.sqrt(cvscores))) + '\n')
    print('  \n Mean time:  ' + str((tf-t0)/5) + '  \n ' )



```
